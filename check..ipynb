```python
import pandas as pd
import numpy as np
import re

# Step 1: Load CSV
file_path = r"D:\scraping_test\data\coe\raw_coe.csv"
df = pd.read_csv(file_path)

# Step 2: Replace specific string with NaN
df['Cost'] = df['Cost'].replace('‡πÑ‡∏°‡πà‡∏û‡∏ö <dt>‡∏Ñ‡πà‡∏≤‡πÉ‡∏ä‡πâ‡∏à‡πà‡∏≤‡∏¢</dt>', np.nan)

# Step 3: Define cost extraction logic
def extract_costs_or_fallback(text):
    if pd.isna(text):
        return np.nan
    
    # Extract numbers (e.g. 25,500, 184000)
    matches = re.findall(r'\d[\d,\.]*', str(text))
    
    if matches:
        return [int(re.sub(r'[,\.]', '', m)) for m in matches]
    
    # If no numbers, check if it looks like a URL
    if re.search(r'https?://', str(text)):
        return text.strip()
    
    # Otherwise return original text
    return text.strip()

# Step 4: Apply function to extract cleaned cost
df['CleanCosts'] = df['Cost'].apply(extract_costs_or_fallback)

# Step 5: Count NaNs
nan_count = df['CleanCosts'].isna().sum()
print(f"‚ùó NaN rows in CleanCosts: {nan_count}")

# Step 6: Get max cost from all lists
all_costs = []
for entry in df['CleanCosts']:
    if isinstance(entry, list):
        all_costs.extend(entry)

if all_costs:
    max_cost = max(all_costs)
    print(f"‚úÖ Max cost found: {max_cost}")
    
    # Replace NaNs with [max_cost]
    df['CleanCosts'] = df['CleanCosts'].apply(
        lambda x: [max_cost] if isinstance(x, float) and np.isnan(x) else x
    )
else:
    print("‚ö†Ô∏è No numeric costs found. Nothing to fill.")
df['term'] = np.nan  # or df['NewColumn'] = ''
df['Total program cost'] = np.nan  # or df['NewColumn'] = ''
# Step 7: Save the result to new CSV
output_path = r"D:\scraping_test\data\coe\cleaned_coe.csv"
df.to_csv(output_path, index=False)
print(f"üìÅ Saved cleaned data to: {output_path}")

df
```

```python
import pandas as pd
import numpy as np
import re

# Step 1: Load CSV
file_path = r"D:\scraping_test\data\aie\raw_aie.csv"
df = pd.read_csv(file_path)

# Step 2: Replace specific string with NaN
df['Cost'] = df['Cost'].replace('‡πÑ‡∏°‡πà‡∏û‡∏ö <dt>‡∏Ñ‡πà‡∏≤‡πÉ‡∏ä‡πâ‡∏à‡πà‡∏≤‡∏¢</dt>', np.nan)

# Step 3: Extract only the maximum number from each row
def extract_max_cost(text):
    if pd.isna(text):
        return np.nan
    
    # Find all number patterns
    matches = re.findall(r'\d[\d,\.]*', str(text))
    
    if matches:
        numbers = [int(re.sub(r'[,\.]', '', m)) for m in matches]
        return max(numbers)  # <- only keep max
    elif re.search(r'https?://', str(text)):
        return text.strip()
    else:
        return text.strip()

# Step 4: Apply to column
df['CleanCosts'] = df['Cost'].apply(extract_max_cost)

# ‚úÖ ADD NEW EMPTY COLUMN HERE
df['term'] = np.nan  # or df['NewColumn'] = ''
df['Total program cost'] = np.nan  # or df['NewColumn'] = ''

# Step 5: Fill NaNs with global max if exists
numeric_costs = df['CleanCosts'].apply(lambda x: x if isinstance(x, (int, float)) else np.nan)
if numeric_costs.notna().any():
    global_max = numeric_costs.max()
    print(f"‚úÖ Global max cost: {global_max}")
    df['CleanCosts'] = df['CleanCosts'].apply(
        lambda x: global_max if pd.isna(x) else x
    )
else:
    print("‚ö†Ô∏è No numeric costs found. Skipping NaN replacement.")

# Step 6: Save
output_path = r"D:\scraping_test\data\aie\cleaned_aie.csv"
df.to_csv(output_path, index=False)
print(f"üìÅ Saved cleaned data to: {output_path}")

# Show result
df
```

```python
import pandas as pd

# Load your original cleaned CSV
# df = pd.read_csv("cleaned_coe.csv")

# Function to determine term and total cost
def extract_or_assume_costs(row):
    cost_text = str(row["Cost"])
    try:
        costs = eval(row["CleanCosts"]) if isinstance(row["CleanCosts"], str) else row["CleanCosts"]
        if not costs or not isinstance(costs, list):
            return pd.Series([row['term'], row['Total program cost']])
        max_cost = max(costs)
        if "‡∏ï‡πà‡∏≠‡∏†‡∏≤‡∏Ñ‡πÄ‡∏£‡∏µ‡∏¢‡∏ô" in cost_text or "‡∏ö‡∏≤‡∏ó/‡πÄ‡∏ó‡∏≠‡∏°" in cost_text:
            return pd.Series([max_cost, max_cost * 8])
        elif "‡∏ï‡πà‡∏≠‡∏†‡∏≤‡∏Ñ‡πÄ‡∏£‡∏µ‡∏¢‡∏ô" not in cost_text and "‡∏ö‡∏≤‡∏ó/‡πÄ‡∏ó‡∏≠‡∏°" not in cost_text:
            return pd.Series([round(max_cost / 8), max_cost])
    except Exception:
        pass
    return pd.Series([row['term'], row['Total program cost']])

# Apply to DataFrame
df[['term', 'Total program cost']] = df.apply(extract_or_assume_costs, axis=1)

# Save the updated file
df.to_csv("coe_with_term_and_total.csv", index=False)
df
```
